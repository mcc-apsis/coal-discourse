```python
import matplotlib.pylab as plt
import numpy as np
import pandas as pd
import time
import sys
import os

import seaborn as sns
import matplotlib.dates as mdates
from datetime import datetime, date
```

```python
# import spread sheet
filename = "Policy events coal.csv"
polevents = pd.read_csv(filename, header=1).fillna("")
polevents
```

```python
# export as latex table

pd.set_option('max_colwidth', 500)

fname = "Policy_events_table.tex"

with open(fname,'w') as tf:
    tf.write(polevents.to_latex(index=False,
                         column_format='rrL{3cm}L{5cm}L{3cm}l', longtable=True).replace('\n',
                                                '\n\\caption{List of events with relation to energy and coal policy.}'
                                                '\n\label{tab:policy-event-list}\n\small\\\\\n', 1))
```
